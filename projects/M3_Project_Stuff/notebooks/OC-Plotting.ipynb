# Load Data

```python
names_list = ['V001', 'V003','V006', 'V007', 'V009','V010', 'V011', 'V012']

data_list = []
for i in range(len(names_list)):
    name = names_list[i]
    data_i = pd.read_csv(name + '.csv')
    data_list.append(data_i)

#data_list
```

```python
x_0 = list(data_list[0]['MJD'])
x_1 = list(data_list[1]['MJD'])
x_2 = list(data_list[2]['MJD'])
x_3 = list(data_list[3]['MJD'])
x_4 = list(data_list[4]['MJD'])
x_5 = list(data_list[5]['MJD'])
x_6 = list(data_list[6]['MJD'])
x_7 = list(data_list[7]['MJD'])

y_0 = list(data_list[0]['O-C'])
y_1 = list(data_list[1]['O-C'])
y_2 = list(data_list[2]['O-C'])
y_3 = list(data_list[3]['O-C'])
y_4 = list(data_list[4]['O-C'])
y_5 = list(data_list[5]['O-C'])
y_6 = list(data_list[6]['O-C'])
y_7 = list(data_list[7]['O-C'])

y_err_0 = list(data_list[0]['O-C err'])
y_err_1 = list(data_list[1]['O-C err'])
y_err_2 = list(data_list[2]['O-C err'])
y_err_3 = list(data_list[3]['O-C err'])
y_err_4 = list(data_list[4]['O-C err'])
y_err_5 = list(data_list[5]['O-C err'])
y_err_6 = list(data_list[6]['O-C err'])
y_err_7 = list(data_list[7]['O-C err'])
```

# Plot Data

```python
def plot_glow(x_data,y_data,y_data_err, ax_i):
    # Redraw the data with low alpha and slighty increased linewidth:
#     n_shades = 10
#     diff_linewidth = 1.05
#     alpha_value = 0.3 / n_shades
#     for n in range(1, n_shades+1):
#         if n == n_shades:
#             ax_i.scatter(x_data,y_data,marker='o',
#                 linewidth=1+(diff_linewidth*n),
#                 alpha=alpha_value,
#                 color='red')

#         ax_i.scatter(x_data,y_data,marker='o',
#                 linewidth=1+(diff_linewidth*n),
#                 alpha=alpha_value,
#                 color='white')
    ax_i.errorbar(x_data,y_data,y_data_err,color='white', fmt='o', ms=6, ecolor='limegreen', capsize=4)
    ax_i.errorbar(x_data[-1],y_data[-1],y_data_err[-1],color='red', fmt='o', ms=6, ecolor='limegreen', capsize=4)



#Generating the very large figure
#one figure with 2 columns, and 4 rows
f, ((ax0,ax1),(ax2,ax3),(ax4,ax5), (ax6,ax7)) = plt.subplots(4,2, figsize=(14,16))

f.suptitle('O-C Diagrams of RR Lyraes in M3')

#top left

ax0.set_title(names_list[0])
ax0.set_xlabel("MJD")
ax0.set_ylabel("O-C")
plot_glow(x_0,y_0,y_err_0, ax0)

#top right

ax1.set_title(names_list[1])
plot_glow(x_1,y_1,y_err_1, ax1)
ax1.set_xlabel("MJD")
ax1.set_ylabel("O-C")

#middle left

ax2.set_title(names_list[2])
plot_glow(x_2,y_2,y_err_2, ax2)
ax2.set_xlabel("MJD")
ax2.set_ylabel("O-C")

#middle right

ax3.set_title(names_list[3])
plot_glow(x_3,y_3,y_err_3, ax3)
ax3.set_xlabel("MJD")
ax3.set_ylabel("O-C")
ax3.set_ylim(-0.1,0.1)

#middle left

ax4.set_title(names_list[4])
plot_glow(x_4,y_4,y_err_4, ax4)
ax4.set_xlabel("MJD")
ax4.set_ylabel("O-C")

#middle right

ax5.set_title(names_list[5])
plot_glow(x_5,y_5,y_err_5, ax5)
ax5.set_xlabel("MJD")
ax5.set_ylabel("O-C")

#bottom left

ax6.set_title(names_list[6])
plot_glow(x_6,y_6,y_err_6, ax6)
ax6.set_xlabel("MJD")
ax6.set_ylabel("O-C")

#bottom right

ax7.set_title(names_list[7])
plot_glow(x_7,y_7,y_err_7, ax7)
ax7.set_xlabel("MJD")
ax7.set_ylabel("O-C")
f.tight_layout()
#plt.savefig('all_OC_diagrams.png', format='png', dpi=300, bbox_inches='tight')
plt.show()
```
