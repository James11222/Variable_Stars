```python
names_list = ['V001', 'V003', 'V005','V006', 'V007', 'V008', 'V009','V010', 'V011', 'V012']

yerr_max_list = [0,0, 0.02, 0.03, 0.03, 0.07,0, 0.05,0,0.05]




jurcsik = pd.read_csv('jurcsik.txt',delim_whitespace=True)
jurcsik.head()
corwin  = pd.read_csv('corwin.txt',delim_whitespace=True,skiprows=13)
corwin.head()
```

# V001 OC Diagram

```python
name = names_list[0] #--------------CHANGE HERE--------------#
band = 'V'
LPP_datafile = 'M3_data/m3_' + names_list[0] + '.dat' #--------------CHANGE HERE--------------#

#-------------
# Corwin data
#-------------
corwin_name = 'v1' #--------------CHANGE HERE--------------#

cut = (corwin['name']==corwin_name) & (corwin['filter']==band)
x = corwin[cut]['JD'].to_numpy() - 2400000
y = corwin[cut]['mag'].to_numpy()
yerr = 0.01 * np.ones_like(y)
star_C01 = PIPS.photdata([x,y,yerr],label='Corwin')

#--------------
# Jurcsik data
#--------------
jurcsik_name = names_list[0] #--------------CHANGE HERE--------------#

cut = (jurcsik['#star']==jurcsik_name) & (jurcsik['band']==band)
x = jurcsik[cut]['HJD-2400000'].to_numpy()
y = jurcsik[cut]['mag'].to_numpy()
yerr = 0.01 * np.ones_like(y)
star_J11 = PIPS.photdata([x,y,yerr],label='Jurcsik')

#--------------------------
# initialize PIPS photdata
#--------------------------
data = PIPS.data_readin_LPP(LPP_datafile,filter=band)
star_KAIT = PIPS.photdata(data,label='KAIT')
# yerr_max_ = yerr_max_list[i]
# star_KAIT.cut(yerr_max=yerr_max_)

#--------------------------
# initialize OC object
#--------------------------
photdata_list = [star_C01,star_J11,star_KAIT]
oc = OC.longdata(photdata_list)

#--------------------------
# run OC! (takes in **kwargs for star.get_period(**kwargs))
#--------------------------
oc.run_OC(p_ref=list(dataFrame_list[0]['mean_period'])[0],template_idx=0,Nterms=7,p_min=0.2,p_max=0.9)

#--------------CHANGE HERE--------------#

#Nterms = 7 for RRab, Nterms = 5 for RRc stars
#We set template_idx to be 0 so Corwin is the reference template

#-------------
# results
#-------------
# outputs
print('O-C:',[float(f'{val:.4f}') for val in oc.oc])      # O-C values
print('err:',[float(f'{val:.4f}') for val in oc.oc_err])  # O-C uncertainties
print('MJD:',[float(f'{val:.1f}') for val in oc.x_means]) # mean x-values
print(f'p_ref: {oc.p_ref:.6f}')   # 'mean' period

# check:
# this value should agree with the offset in O-C values 
# between Jurcsik and Corwin.

# correct for rounding ('round','floor','ceil',etc.)
offset_JC = oc.oc[1]-oc.oc[0]
if offset_JC > oc.p_ref/2:
    offset_JC -= oc.p_ref
elif offset_JC < -oc.p_ref/2:
    offset_JC += oc.p_ref
    
print(f'Jurcsik - Corwin: {offset_JC:.5f}')

# the result:
# this is the OC-value offset between Jurcsik and KAIT!
# (simply add this to Jurcsik's O-C value to get ours in the same frame)
print(f'KAIT - Jurcsik: {oc.oc[2]-oc.oc[1]:.5f}') 

# correct for rounding ('round','floor','ceil',etc.)
offset_KAIT = oc.oc[2]-oc.oc[1]
if offset_KAIT > oc.p_ref/2:
    offset_KAIT -= oc.p_ref
elif offset_KAIT < -oc.p_ref/2:
    offset_KAIT += oc.p_ref

# this is the OC-value offset between Jurcsik and KAIT!
# (simply add this to Jurcsik's O-C value to get ours in the same frame)
OC_VALUE = list(dataFrame_list[0]['O-C'][:])[-1] + offset_KAIT #--------------CHANGE HERE--------------#
OC_err = np.sqrt((oc.oc_err[2]-oc.oc_err[1])**2 + (list(dataFrame_list[0]['O-C err'][:])[-1])**2)
#--------------CHANGE HERE--------------#

# if OC_VALUE > oc.p_ref/2:
#     OC_VALUE -= oc.p_ref
# elif OC_VALUE < -oc.p_ref/2:
#     OC_VALUE += oc.p_ref


print('the MJD is {0:0.8f} and the O-C Value is {1:0.8f} +/- {2:0.8f}'.format(oc.x_means[2], OC_VALUE, OC_err))
```

```python
x_data = np.array((dataFrame_list[0]['mean HJD'][:])) - 2400000
y_data = np.array((dataFrame_list[0]['O-C'][:]))
y_data_err = np.array((dataFrame_list[0]['O-C err'][:]))
x_, y_, y_err  = oc.x_means[2], OC_VALUE, OC_err
 

x_data = np.append(x_data,x_)
y_data = np.append(y_data,y_)
y_data_err = np.append(y_data_err, y_err)

df_ = pd.DataFrame({'MJD':x_data,'O-C':y_data,'O-C err':y_data_err})
df_.to_csv(jurcsik_name + '.csv')


# #Pretty Plot
# plt.figure(figsize=(8,4))
# ax = plt.gca()
# plt.title(jurcsik_name)
# plt.xlabel("MJD")
# plt.ylabel("O-C")

# # Redraw the data with low alpha and slighty increased linewidth:
# n_shades = 10
# diff_linewidth = 1.05
# alpha_value = 0.3 / n_shades
# for n in range(1, n_shades+1):
#     if n == n_shades:
#         plt.scatter(x_data,y_data,marker='o',
#             linewidth=1+(diff_linewidth*n),
#             alpha=alpha_value,
#             color='red')
        
#     plt.scatter(x_data,y_data,marker='o',
#             linewidth=1+(diff_linewidth*n),
#             alpha=alpha_value,
#             color='white')
# plt.errorbar(x_data,y_data,y_data_err,color='white', fmt='o', ms=6, ecolor='limegreen', capsize=4)
# plt.errorbar(x_,y_,y_err,color='red', fmt='o', ms=6, ecolor='limegreen', capsize=4)
# #plt.scatter(x_,y_,color='red', marker='o')
# #plt.grid()
# #plt.savefig('OC_VOO1.png', format='png', dpi=300, bbox_inches='tight')
# plt.show()
```
